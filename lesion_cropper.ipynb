```python
import skimage.io as io
import numpy as np
from skimage.transform import resize
import matplotlib.pyplot as plt

img=io.imread("C:/Users/mihae/Documents/GitHub/mdasm-2024/pictures/PAT_386_785_536.png")
mask=io.imread("C:/Users/mihae/Documents/GitHub/mdasm-2024/groupR_masks/PAT_386_785_536_mask.png")
def lesion_cropper(img, mask):
    object=np.where(mask!=0)
    min_x=min(object[0])
    max_x=max(object[0])
    min_y=min(object[1])
    max_y=max(object[1])
    cropped=img[min_x-5:max_x+5, min_y-5:max_y+5]
    cropped_mask=mask[min_x-5:max_x+5, min_y-5:max_y+5]
    size=max(max_x-min_x, max_y-min_y)
    multiplier=500/size
    new_img=resize(cropped, (cropped.shape[0] * multiplier, cropped.shape[1] * multiplier), anti_aliasing=False)
    new_mask=resize(cropped_mask, (cropped_mask.shape[0] * multiplier, cropped_mask.shape[1] * multiplier), anti_aliasing=False)
    plt.imshow(new_img)
    plt.imshow(new_mask)

lesion_cropper(img, mask)
#plt.imshow(mask)
```
